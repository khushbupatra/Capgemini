```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
from sklearn.datasets import load_breast_cancer,fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,r2_score,mean_absolute_error
```

```python
cdata=load_breast_cancer()
```

```python
cdata
```

```python
df=pd.DataFrame(cdata.data,columns=cdata.feature_names)
df['cancer']=cdata.target
```

```python
df
```

```python
x=df.drop('cancer',axis=1)
y=df['cancer']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
##n_ngeighbors are the k value so here we have taken k as 10 
model=KNeighborsClassifier(n_neighbors=10)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
con_mat = confusion_matrix(y_test, y_pred)
con_mat
```

```python
print('Training Efiiciency: ', model.score(x_train, y_train))
print('Testing Efiiciency: ', model.score(x_test, y_test))
```

KNN Regressor

```python
rdata = fetch_california_housing()
rdata
```

```python
df2 = pd.DataFrame(rdata.data, columns=rdata.feature_names)
df2['price'] = rdata.target
df2
```

```python
X = df2.drop('price', axis = 1)
Y = df2['price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state= 42)
```

```python
model2 = KNeighborsRegressor(n_neighbors = 3)
model2.fit(X_train, Y_train)
```

```python
y_pred2 = model2.predict(X_test)
y_pred2
```

```python
print('Training Efiiciency: ', model2.score(X_train, Y_train))
print('Testing Efiiciency: ', model2.score(X_test, Y_test))
```

```python
mae = mean_absolute_error(Y_test, y_pred2)
mae
```
