```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv')
```

```python
df.head()
```

```python
df.info()
```

Quality Column


```python
pop=df['quality']
pop
```

```python
sample_size=1000
sample=df['quality'].sample(sample_size)
```

```python
sample
```

```python
sample_mean=sample.mean()
sample_mean
```

```python
pop_mean=pop.mean()
pop_mean
```

```python
error=pop_mean-sample_mean
```

```python
error
```

```python
##scatter_plot
plt.scatter(range(1,sample_size+1),sample)
plt.axhline(pop.mean(),color='red',label='population mean')
plt.axhline(sample.mean(),color='blue',label='sample mean')
plt.legend()
plt.show()
```
