```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
from sklearn.datasets import load_breast_cancer,fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,r2_score,mean_absolute_error
```

```python
cdata=load_breast_cancer()
```

```python
cdata
```

```python
df=pd.DataFrame(cdata.data,columns=cdata.feature_names)
df['cancer']=cdata.target
```

```python
df
```

```python
x=df.drop('cancer',axis=1)
y=df['cancer']
```

```python
x_train,x_test,y_test,y_train=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
##n_ngeighbors are the k value so here we have taken k as 10 
model=KNeighborsClassifier(n_neighbors=10)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```
