```python
##write a logic to remove all the duplicate from a list 
list = [1,2,3,4,1,2,3,4,5]
without_dup=set(list)
print(without_dup)
```

```python
## write a logic to remove dulpicates form a list 
list = [1,2,3,4,1,2,3,4,5]
without_dup=[]
for i in list:
    if i not in without_dup:
        without_dup+=[i]
without_dup    
```
