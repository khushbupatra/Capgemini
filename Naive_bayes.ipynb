```python
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
```

```python
data=load_diabetes()
df=pd.DataFrame(data.data,columns=data.feature_names)
df['diabetes']=data.target
df.head()
```

```python
df['diabetes'].dtype
```

```python
out=[]
for i in df.diabetes:
    if i >= 120.0:
        out.append(1)
    else:
        out.append(0)
df['diabetes']=out
df['diabetes'].value_counts()            
```

```python
df
```

```python
df['diabetes'].value_counts()
```

```python
x=df.drop('diabetes',axis=1)
y=df.diabetes
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.7,random_state=42)
```

```python
model=GaussianNB()
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
print('training efficiency',model.score(xtrain,ytrain))
print('testing efficiency',model.score(xtest,ytest))
```

```python
print('Accuracy score: ', accuracy_score(ytest, ypred))
print('Precision score: ', precision_score(ytest, ypred))
print('Recall score: ', recall_score(ytest, ypred))
print('F1 score: ', f1_score(ytest, ypred))
```
