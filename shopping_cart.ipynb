```python
class shoppingcart:
    products={'iphone':5,'imac':3,'ipad':2,'iwatch':1}
    prices={'iphone':900,'imac':5000,'ipad':3000,'iwatch':4000}

    def __init__(self):
        self.cart=[]
        ##this is constructor which will be called when we create object of shoppingcart class
        ##it will initialize the cart as empty list
        ##we can add items to the cart using add_items method and remove items using remove_items method and calculate total cost using total_cost method

    def add_items(self,name,quantity):
        if name not in shoppingcart.products.keys():
            raise Exception(f'cannot add products{name}')
        if quantity>shoppingcart.products[name]: 
            raise Exception(f'out of stock')
        item={'name':name,'quantity':quantity,'price':shoppingcart.prices[name]}
        self.cart.append(item)
        shoppingcart.products[name]-=quantity

    def remove_items(self,name):
        for item in self.cart:
            if item['name']==name:
                if item['quantity']==1:
                    self.cart.remove(item)
                else:
                    item['quantity']=item['quantity']-1

    def total_cost(self):
        total=0
        for item in self.cart:
            total+=item['quantity']*item['price']
        return total
    
cart1=shoppingcart()
cart2=shoppingcart()
cart1.add_items('iphone',2)
cart1.add_items('imac',1)
cart2.add_items('ipad',1)
print(cart1.cart)
print(cart2.cart)
```
